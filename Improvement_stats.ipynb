```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dates = ['31.07.2023','05.08.2023' ]
metrics = [
    [21.214285714285715,
    18.428571428571427,
    0.42857142857142855,
    8.4375,
    5.6875,
    16.5],
    [14.607142857142858,
    11.928571428571429,
    0.39285714285714285,
    8.294117647058824,
    5.470588235294118,
    20.058823529411764]
    ]
metrics_np = np.array(metrics)
data_series = np.transpose(metrics_np)
data_series
```

```python
for row in data_series[0:2,:]:
  print(row)
  plt.plot(dates, row)
plt.xlabel("Table date")
plt.ylabel("Performance")
plt.title("All test cases")
plt.legend(['Rank 10 (mean)', 'Rank 25 (mean)'])

plt.show()
```

```python
print(data_series[2,:])
plt.plot(dates, data_series[2,:])
plt.xlabel("Table date")
plt.ylabel("Performance")
plt.title("Converging cases")
plt.legend(['Convergence %'])

plt.show()
```

```python
for row in data_series[3:,:]:
  print(row)
  plt.plot(dates, row)
plt.xlabel("Table date")
plt.ylabel("Performance")
plt.title("Converging test cases")
plt.legend(['Rank 10 (mean)', 'Rank 25 (mean)', 'Rank converged (mean)'])

plt.show()
```
